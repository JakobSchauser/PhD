```python
import numpy as np
```

```python
fut1 = np.genfromtxt('fut1.csv', delimiter=',', dtype=str, skip_header=0)

header = fut1[0]
print(header[2:12])

fut1 = fut1[:,2:12]
```

```python
fut2 = np.genfromtxt('fut2.csv', delimiter=',', dtype=str, skip_header=0)

header2 = fut2[0]
print(header2[2:9])
fut2 = fut2[:,2:9]
```

```python
fut3 = []

i = 0
for line in open('fut3.csv'):
    if i == 0:
        i += 1
        continue
    fut3.append(line.strip().split(',')[2:14])  

fut3 = np.array(fut3)

header3 = fut3[0]
print(header3)
```

```python
fut4 = np.genfromtxt('fut4.csv', delimiter=',', dtype=str, skip_header=0)

header4 = fut4[0]
print(header4[2:13])
fut4 = fut4[:,2:13]

```

```python
# save to cleaned files
np.savetxt('cleanfut1.csv', fut1, delimiter=',', fmt='%s')
np.savetxt('cleanfut2.csv', fut2, delimiter=',', fmt='%s')
np.savetxt('cleanfut3.csv', fut3, delimiter=',', fmt='%s')
np.savetxt('cleanfut4.csv', fut4, delimiter=',', fmt='%s')
```
